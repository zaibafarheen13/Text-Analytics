```python
import nltk
from nltk import word_tokenize
from nltk.probability import FreqDist
import urllib.request
from wordcloud import WordCloud
nltk.download('punkt')
%matplotlib inline
```

```python
textFile=open('C:/Users/Zaiba Farheen/Downloads/pdfcoffee.com_crooked-house-by-agatha-christie-pdf-free.txt','r').read()
#print(textFile)
print(textFile[500:1200])
```

```python
#tokenize text by words
words=word_tokenize(textFile)
print(words)
```

```python
#check the number of words
print(f"The total number of words in the text is {len(words)}")
```

```python
#find the frequency of words
fdist=FreqDist(words)
#print 10 most commmon words
fdist.most_common(10)
```

```python
#create an empty list to store words
words_no_punc=[]

#iterate through word list to remove punctuation and numbers
for word in words:
    if word.isalpha():
        words_no_punc.append(word.lower())

#print number of words without punctuation
print(f"The total number of words without punctuation is {len(words_no_punc)}")
```

```python
#Download and import list of stopwords
nltk.download("stopwords")
from nltk.corpus import stopwords

#list of stopwords
stopwords_list=stopwords.words("english")
print(stopwords_list)
```

```python
#clean an empty list to store clean words
clean_words=[]

#iterate through the words_no_punc list and add non stopwords to the new clean_
for word in words_no_punc:
    if word not in stopwords_list:
        clean_words.append(word)
        
print(f"The total number of words without punctuation and stopwords is {len(clean_words)}")
print(clean_words)
```

```python
#find the frequency of words
fdist=FreqDist(clean_words)
#print 10 most commmon words
fdist.most_common(10)
```
